```python
clicks_per_app=frame.groupby('app_bundle')['app_bundle'].count().reset_index(name="count")
conversions_per_app=frame.groupby('app_bundle')['converted'].sum().reset_index(name="converted")

ratingdf=pd.merge(clicks_per_app,conversions_per_app,on='app_bundle')

ratingdf['app_rating']=100*ratingdf['converted']/ratingdf['count']
ratingdf=ratingdf[['app_bundle','app_rating']]
extframe =pd.merge(frame,ratingdf,on='app_bundle')

clicks_per_user=frame.groupby('device_ifa')['device_ifa'].count().reset_index(name="count")
conversions_per_user=frame.groupby('device_ifa')['converted'].sum().reset_index(name="converted")

consumptiondf=pd.merge(clicks_per_user,conversions_per_user,on='device_ifa')

consumptiondf['user_consumption']=100*consumptiondf['converted']/consumptiondf['count']
consumptiondf=consumptiondf[['device_ifa','user_consumption']]
extframe =pd.merge(extframe,consumptiondf,on='device_ifa')
```
