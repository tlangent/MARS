```python
clicks_per_app=frame.groupby('app_bundle')['app_bundle'].count().reset_index(name="count")
```

```python
conversions_per_app=frame.groupby('app_bundle')['converted'].sum().reset_index(name="conversions")
```

```python
clicks_per_app=frame.groupby('app_bundle')['app_bundle'].count().reset_index(name="app_clicks")
conversions_per_app=frame.groupby('app_bundle')['converted'].sum().reset_index(name="app_conversions")

ratingdf=pd.merge(clicks_per_app,conversions_per_app,on='app_bundle')

ratingdf['app_rating']=100*ratingdf['app_conversions']/ratingdf['app_clicks']
#ratingdf=ratingdf[['app_bundle','app_rating']]
extframe =pd.merge(frame,ratingdf,on='app_bundle')

clicks_per_user=frame.groupby('device_ifa')['device_ifa'].count().reset_index(name="clicks")
conversions_per_user=frame.groupby('device_ifa')['converted'].sum().reset_index(name="conversions")

consumptiondf=pd.merge(clicks_per_user,conversions_per_user,on='device_ifa')

consumptiondf['user_consumption']=100*consumptiondf['conversions']/consumptiondf['clicks']
consumptiondf=consumptiondf[['device_ifa','user_consumption']]
extframe =pd.merge(extframe,consumptiondf,on='device_ifa')
```
